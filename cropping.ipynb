## Cropping

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("./images/input.jpg")
plt.imshow(image[..., ::-1])
```

```python
height, width = image.shape[:2]

cropped = image[280:405, 287:533]

plt.imshow(cropped[..., ::-1])

cv.imwrite("cropped.png", cropped)
cv.imwrite("cropped.jpg", cropped)
```

```python
cv.imshow("image", image)
cv.imshow("cropped", cropped)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
height, width = image.shape[0:2]

cropped = image[height//3:3*height//4,width//3:3*width//4]

plt.imshow(cropped[..., ::-1])
```
