```python
m1 = np.array([[150, 155], [156, 157]], dtype=np.uint8)
m2 = np.ones([2, 2], dtype=np.uint8) * 100
print("m1:")
print(m1)
print("m2:")
print(m2)
print("m1 + m2:")
print(m1 + m2)
```

```python
m1 = np.array([[150, 155], [156, 157]], dtype=np.uint8)
m2 = np.ones([2, 2], dtype=np.uint8) * 100

modified_m1 = m1.astype("int16") + m2
modified_m1 = np.minimum(modified_m1, 255)
modified_m1 = modified_m1.astype(np.uint8)
modified_m1
```

```python
m1 = np.array([[150, 155], [156, 157]], dtype=np.uint8)

value_to_add = 100
max_threshold = 255 - value_to_add

m1[m1 >= max_threshold] = 255
m1[m1 < max_threshold] += value_to_add

print(m1)
```

```python
image = cv.imread("./images/input.jpg")

M = np.ones(image.shape, dtype=np.uint8) * 100

added = cv.add(image, M)
subtracted = cv.subtract(image, M)

plt.figure(figsize=(15, 10))
plt.subplot(131);
plt.imshow(image[..., ::-1]);
plt.title("Original")
plt.subplot(132);
plt.imshow(added[..., ::-1]);
plt.title("Added")
plt.subplot(133);
plt.imshow(subtracted[..., ::-1]);
plt.title("Subtracted")

cv.imshow("Original", image)
cv.imshow("Added", added)
cv.imshow("Subtracted", subtracted)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img1 = cv.imread("./images/class.vision.jpg")
shape = img1.shape
img2 = cv.imread("./images/input.jpg")[100:100 + shape[0], 100:100 + shape[1], :]

output_image = cv.add(img1, img2)

plt.figure(figsize=(10, 5))
plt.subplot(131);
plt.imshow(img1[..., ::-1]);
plt.title("Image1")
plt.subplot(132);
plt.imshow(img2[..., ::-1]);
plt.title("Image2")
plt.subplot(133);
plt.imshow(output_image[..., ::-1]);
plt.title("Image1 + Image2")

```

```python
img1 = cv.imread("./images/input.jpg")
img2 = cv.imread("./images/class.vision.jpg")

print(img1.shape)
print(img2.shape)
```

```python
logo_h, logo_w, _ = img2.shape
location_w, location_h = 500, 10
padded_img2 = np.zeros_like(img1)
padded_img2[location_h:location_h + logo_h, location_w:location_w + logo_w] = img2
plt.imshow(padded_img2[..., ::-1])
```

```python
output_image = cv.add(img1, padded_img2)
plt.imshow(output_image[..., ::-1])
```

```python
img1 = cv.imread("./images/opencv.jpg")
img2 = cv.imread("./images/class.vision.jpg")
output_image = cv.addWeighted(img1, 0.3, img2, 0.7, 0)

plt.figure(figsize=(10, 5))
plt.subplot(131)
plt.imshow(img1[..., ::-1])
plt.title("Image1")
plt.subplot(132)
plt.imshow(img2[..., ::-1])
plt.title("Image2")
plt.subplot(133)
plt.imshow(output_image[..., ::-1])
plt.title("Weighted sum")
```

```python
output_image2_8 = cv.addWeighted(img1, 0.2, img2, 0.8, 0)
output_image5_5 = cv.addWeighted(img1, 0.5, img2, 0.5, 0)
output_image8_2 = cv.addWeighted(img1, 0.8, img2, 0.2, 0)

plt.figure(figsize=(10, 5))
plt.subplot(131)
plt.imshow(output_image2_8[..., ::-1])
plt.title("Output image2_8")
plt.subplot(132)
plt.imshow(output_image5_5[..., ::-1])
plt.title("Output image5_5")
plt.subplot(133)
plt.imshow(output_image8_2[..., ::-1])
plt.title("Output image8_2")
```
