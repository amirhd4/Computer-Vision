## How to convert colored image to grayscale

```python
import cv2 as cv

image = cv.imread("./images/input.jpg")
cv.imshow("input", image)
cv.waitKey()

gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

cv.imshow("Grayscale", gray_image)
cv.waitKey()
cv.destroyAllWindows()
```

```python
img = cv.imread("./images/input.jpg", 0)
cv.imshow("Grayscale", img)
cv.waitKey()
cv.destroyAllWindows()
```
