```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
square = np.zeros((300, 300), np.uint8)
cv.rectangle(square, (50, 50), (250, 250), 250, -1)

ellipse = np.zeros((300, 300), np.uint8)
cv.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)

plt.subplot(121);plt.imshow(square, cmap='gray');plt.title('square')
plt.subplot(122);plt.imshow(ellipse, cmap='gray');plt.title('ellipse')
```

```python
bitwise_and = cv.bitwise_and(square, ellipse)

bitwise_or = cv.bitwise_or(square, ellipse)

bitwise_xor = cv.bitwise_xor(square, ellipse)

bitwise_not = cv.bitwise_not(square)

plt.figure(figsize=(12, 6))
plt.subplot(242);plt.imshow(square, cmap='gray');plt.title('Square')
plt.subplot(243);plt.imshow(ellipse, cmap='gray');plt.title('Ellipse')
plt.subplot(245);plt.imshow(bitwise_and, cmap='gray');plt.title('Bitwise AND')
plt.subplot(246);plt.imshow(bitwise_or, cmap='gray');plt.title('Bitwise OR')
plt.subplot(247);plt.imshow(bitwise_xor, cmap='gray');plt.title('Bitwise XOR')
plt.subplot(248);plt.imshow(bitwise_not, cmap='gray');plt.title('Bitwise NOT')
```

```python
image = cv.imread("./images/input.jpg")
cropped = image[100:600, 150:650]

circle = np.zeros((500, 500, 3), "uint8")
cv.circle(circle, (250, 250), 250, (255, 255, 255), -1)

output_image = cv.bitwise_and(cropped, circle)
plt.imshow(output_image[...,::-1], cmap='gray')
```

```python
png_image = np.zeros((500, 500, 4), np.uint8)
png_image[...,0:3] = cropped
circle = np.zeros((500, 500), np.uint8)
cv.circle(circle, (250, 250), 250, (128, 128, 128), -1)
png_image[..., 3] = circle
cv.imwrite("caw.png", png_image)
```

```python
square = np.zeros((10, 10), np.uint8)
cv.rectangle(square, (2, 2), (8, 8), 2, -1)

circle1 = np.zeros((10, 10), np.uint8)
cv.circle(circle1, (5, 5), 2, 3, -1)

circle2 = np.zeros((10, 10), np.uint8)
cv.circle(circle2, (5, 5), 2, 4, -1)

output1 = cv.bitwise_and(square, circle1)
output2 = cv.bitwise_and(square, circle2)

print("---square---")
print(square)
print("--circle1---")
print(circle1)
print("--circle2---")
print(circle2)
print("---output1---")
print(output1)
print("---output2---")
print(output2)
```
